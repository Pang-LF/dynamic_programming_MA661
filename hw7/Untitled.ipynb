```python
def v(t, x, o):
    if t==18 and x==0:
        vt=0
    elif t==18 and x==1:
        vt=2
    elif t<18 and x==0:
        vt=0.8**o*v(t+1,x,o+1)+0.8**(o-1)*0.2*v(t+1,x+1,1)
    elif t<18 and x==1:
        vt=min(4+v(t+1,x-1,1), 2+v(t+1,x,o+1))
    return vt

i=17
j=1
while i>0:
    while j<=i:
        print('v', i, '(1,', j, ')=', v(i,0,j))
        print('v', i, '(2,', j, ')=', v(i,1,j))
        j+=1
    j=1
    i-=1
```
